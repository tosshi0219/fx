```python
from datetime import datetime, timedelta
import numpy as np
import oandapy
import sys
sys.path.append('/Users/toshio/project/fx')
from config import *
from lib.inference import Inference
from lib.trader import Trader
import time
```

```python
ENV = 'practice'
oanda = oandapy.API(environment=ENV, access_token=token)
```

```python
trader = Trader()
volume = 100
set_profit = 0.1
set_loss = 0.5
```

```python
#100時間取引する
j = 0
while j < 1000:
    j +=1
    now = datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S')
    print('------transaction_{0:03}------'.format(j))
    print(now)
    inf = Inference()
    pred_delta=float(inf.pred[0][0])
    pred_side = inf.pred_side
    print('1時間後の推定差額は{}'.format(np.round(pred_delta, 4)))
    print(pred_side)
    #現在価格の取得
    current_price = oanda.get_prices(instruments='USD_JPY')['prices'][0]
    current_ask_price = current_price['ask']
    current_bid_price = current_price['bid']

    #現在のポジション確認
    open_trades = oanda.get_trades(account_id)
    if len(open_trades['trades']) == 0:
        current_side = 'zero'
    else:
        current_side = open_trades['trades'][0]['side']
        #最も古い取引からの経過時間の確認
        trade_time = datetime.strptime(open_trades['trades'][0]['time'], '%Y-%m-%dT%H:%M:%S.000000Z')
        current_time = datetime.strptime(now,'%Y-%m-%d %H:%M:%S')
        elapsed_time = (current_time - trade_time).seconds/3600

    if (current_side == 'zero') & (pred_side == 'buy'):
        take_profit = current_ask_price + set_profit
        stop_loss = current_bid_price - set_loss
        trader._market_order(pred_side, volume, take_profit, stop_loss)
        print("Let's buy!")
    elif (current_side == 'zero') & (pred_side == 'sell'):
        take_profit = current_bid_price - set_profit
        stop_loss = current_ask_price + set_loss
        trader._market_order(pred_side, volume, take_profit, stop_loss)
        print("Let's sell!")
    elif (current_side == 'buy') & (pred_side == 'buy'):
        take_profit = current_ask_price + set_profit
        stop_loss = current_bid_price - set_loss
        trader._market_order(pred_side, volume, take_profit, stop_loss)        
        print("Let's buy!")
    elif (current_side == 'buy') & (pred_side != 'buy'):
        if elapsed_time > 1:
            trade_id = open_trades['trades'][0]['id']
            closed_trades = oanda.close_trade(account_id, trade_id)
            profit = closed_trades['profit']
            print('profit:{}'.format(profit))
        else:
            print('do nothing')
            pass
    elif (current_side == 'sell') & (pred_side == 'sell'):
        take_profit = current_bid_price - set_profit
        stop_loss = current_ask_price + set_loss
        trader._market_order(pred_side, volume, take_profit, stop_loss)
        print("Let's sell!")
    elif (current_side == 'sell') & (pred_side != 'sell'):
        if elapsed_time > 1:
            trade_id = open_trades['trades'][0]['id']
            closed_trades = oanda.close_trade(account_id, trade_id)
            profit = closed_trades['profit']
            print('profit:{}'.format(profit))
        else:
            print('do nothing')
            pass
    else:
        print('do nothing')
        pass
      
    print('-------------------------------------')
    time.sleep(5) #5秒に設定中
```
