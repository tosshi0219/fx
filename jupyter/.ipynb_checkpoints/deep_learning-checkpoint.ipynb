### トレンド指標の定義

```python
def moving_average(df, window):
    mov_ave = df.rolling(window).mean().iloc[window-1:,:]    
    
    plt.figure(figsize = (20,5))
    plt.plot(df['closeAsk'])
    plt.plot(mov_ave['closeAsk'])
    plt.xlim(['2005-01-01','2006-01-01'])
    plt.ylim([100,130])
    plt.show()
    return mov_ave
```

```python
def ichimoku(df):
    max9 = df.rolling(9).max()
    min9 = df.rolling(9).min()
    max26 = df.rolling(26).max()
    min26 = df.rolling(26).min()
    max52 = df.rolling(52).max()
    min52 = df.rolling(52).min()
    chikoindex = [df.index[0] +  timedelta(days = a) for a in range(-26,0)]
    chikoadder = list(df.index[:-26])
    chikoindex.extend(chikoadder)
    senkoindex = list(df.index[26:])
    senkoadder = [df.index[-1] +  timedelta(days = a) for a in range(1,27)]
    senkoindex.extend(senkoadder)

    tenkan = (max9 + min9) / 2
    kijun = (max26 + min26) / 2
    chiko = df.copy()
    chiko['time'] = chikoindex
    chiko = chiko.set_index('time', drop = True)
    senko1 = (tenkan + kijun) / 2
    senko2 = (max52 + min52) / 2
    senko1['time'] = senkoindex
    senko1 = senko1.set_index('time', drop = True)
    senko2['time'] = senkoindex
    senko2 = senko2.set_index('time', drop = True)

    ichi =pd.DataFrame()
    col = 'closeAsk'
    ichi['close'] = df[col]
    ichi['tenkan'] = tenkan[col]
    ichi['kijun'] = kijun[col]
    ichi['chiko'] = chiko[col]
    ichi['senko1'] = senko1[col]
    ichi['senko2'] = senko2[col]
    
    plt.figure(figsize = (20,5))
    plt.plot(ichi)
    plt.xlim(['2005-01-01','2006-01-01'])
    plt.ylim([100,130])
    plt.show()
    return ichi
```

```python
def bollinger_band(df, window):
    bband = pd.DataFrame()
    bband['close'] = df['closeAsk']
    bband['mean'] = df['closeAsk'].rolling(window).mean()
    bband['std'] = df['closeAsk'].rolling(window).std()
    bband['upper1'] = bband['mean'] + (bband['std'] * 1)
    bband['lower1'] = bband['mean'] - (bband['std'] * 1)
    bband['upper2'] = bband['mean'] + (bband['std'] * 2)
    bband['lower2'] = bband['mean'] - (bband['std'] * 2)
    bband['upper3'] = bband['mean'] + (bband['std'] * 3)
    bband['lower3'] = bband['mean'] - (bband['std'] * 3)
    
    plt.figure(figsize = (20,5))
    plt.plot(bband['close'])
    plt.plot(bband['mean'])
    plt.plot(bband['upper1'])
    plt.plot(bband['lower1'])
    plt.plot(bband['upper2'])
    plt.plot(bband['lower2'])
    plt.plot(bband['upper3'])
    plt.plot(bband['lower3'])
    plt.xlim(['2005-01-01','2006-01-01'])
    plt.ylim([100,130])
    plt.show()
    return bband
```

```python
mov_ave = moving_average(df, 20)
ichi = ichimoku(df)
bband = bollinger_band(df, 20)
```

### オシレータ指標の定義

```python
def rsi(df):
    diff = pd.DataFrame()
    diff['RSI'] = df['closeAsk'].diff()[1:]
    up, down = diff.copy(), diff.copy()
    up[up < 0] = 0
    down[down > 0] = 0
    up_sma_14 = up.rolling(14).mean()
    down_sma_14 = down.abs().rolling(14).mean()
    RS = up_sma_14 / down_sma_14
    RSI = 100.0 - (100.0 / (1.0 + RS))
    
    plt.figure(figsize = (20,5))
    plt.plot(RSI)
    plt.xlim(['2005-01-01','2006-01-01'])
    plt.grid()
    plt.show()
    return RSI
```

```python
def macd(df):
    MACD = pd.DataFrame()
    MACD['close'] = df['closeAsk']
    MACD['ema_12'] = df['closeAsk'].ewm(span=12).mean()
    MACD['ema_26'] = df['closeAsk'].ewm(span=26).mean()
    MACD['MACD'] = MACD['ema_12'] - MACD['ema_26']
    MACD['signal'] = MACD['MACD'].ewm(span=9).mean()
    
    plt.figure(figsize = (20,5))
    plt.plot(MACD['MACD'])
    plt.plot(MACD['signal'])
    plt.xlim(['2005-01-01','2006-01-01'])
    plt.grid()
    plt.show()
    return MACD
```

```python
RSI = rsi(df)
MACD = macd(df)
```

### データフレームの作成

```python
df1 = df.join(RSI)
df2 = df1.join(MACD[['MACD','signal']])
df3 = df2.join(bband)
df4 = df3.join(ichi.drop('close', axis =1))
df5 = df4.drop('volume', axis = 1)
```

### トレンドライン

```python
def trend_line(df, kijun, window):
    start = kijun - timedelta(days = window)
    df = df.loc[start:kijun,:]
    df['time_id'] = np.arange(1,df.shape[0]+1)
    df_fin = df.copy()
    df_high = df.copy()
    df_low = df.copy()

    while len(df_high)>3:
        reg_1 = linregress(
                        x = df_high['time_id'],
                        y = df_high['highAsk']
                        )
        df_high = df_high.loc[df_high['highAsk'] > reg_1[0] * df_high['time_id'] + reg_1[1]]

    reg_1 = linregress(
                    x = df_high['time_id'],
                    y = df_high['highAsk']
                    )

    df_fin['high_trend'] = reg_1[0] * df_fin['time_id'] + reg_1[1]

    while len(df_low)>3:
        reg_2 = linregress(
                        x = df_low['time_id'],
                        y = df_low['lowAsk']
                        )
        df_low = df_low.loc[df_low['lowAsk'] < reg_2[0] * df_low['time_id'] + reg_2[1]]

    reg_2 = linregress(
                    x = df_low['time_id'],
                    y = df_low['lowAsk']
                    )

    df_fin['low_trend'] = reg_2[0] * df_fin['time_id'] + reg_2[1]

    # トレンドラインをプロット
    plt.figure(figsize = (20,5))
    plt.plot(df_fin['time_id'],df_fin['highAsk'])
    plt.plot(df_fin['time_id'],df_fin['lowAsk'])
    plt.plot(df_fin['time_id'],df_fin['high_trend'])
    plt.plot(df_fin['time_id'],df_fin['low_trend'])
    plt.grid()
    plt.show()
```

```python
kijun = datetime(2018,7,31)
window = 100
trend_line(df, kijun, window)
```
